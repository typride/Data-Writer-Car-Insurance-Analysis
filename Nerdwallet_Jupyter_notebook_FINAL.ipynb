# Now that we have a list of each subset of the data, let's make some visualizations.

The 2 Cells below create a function that generates box plots to visualize data for any paramaters you send it. This can be multi-use, and be tailored to any data that you may have. The box plots that you see below are what we can use to create our articles. I think it could be valuable for our readers to have some of this data, but that can be for another article. This current article will require the medians for each of these, which I have generated after the box plot cells. 

```python
def boxplot_create(df_param, gender, age, name_infraction):
    # Initialize figure 
    fig, ax = plt.subplots(figsize=(12,8))

    # plot boxplots 
    bp = df_param.boxplot(column="AnnualPremium", by="Company", ax=ax, return_type="dict")

    # Set title and vertical axis label
    ax.set_title('Cost of Insurance by Company for ' + gender + ' Aged ' + str(age) + '  w/ 1 ' + name_infraction, fontsize=18)
    ax.set_ylabel("Annual Premium ($)", fontsize=16)

    # Make grid-lines lighter
    ax.grid(alpha=0.25);
```

```python
age_count = 0
list_infractions = ["Speeding Ticket", "Speeding Ticket", "Speeding Ticket", "Speeding Ticket",  "DUI","DUI","DUI","DUI", "Accident","Accident","Accident","Accident","Incident", "Incident","Incident", "Incident"]
for i in range(len(list_data)):
    if(age_count <= 1):
       age = 21
    elif(age_count<= 3):
       age = 35
    else:
       age = 21
    age_count+=1
       
    if i % 2 == 0:
        boxplot_create(list_data[i], 'Male', age, list_infractions[i])
    else:
        boxplot_create(list_data[i], 'Female', age, list_infractions[i])
```
